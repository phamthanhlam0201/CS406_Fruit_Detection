```python
import yaml
import os

mapping = ['apple', 'orange', 'lemon',
           'avocado', 'cherry', 'coconut',
           'banana', 'watermelon', 'pineapple',
           'kiwi', 'tomato', 'mango']

dataset_info = {
    'train': 'dataset/dataset/train/images',
    'val': 'dataset/dataset/val/images',
    'test': 'dataset/dataset/test/images',
    'path': os.path.abspath('/kaggle/input'),
    'nc': len(mapping),
    'names': mapping
}

yaml_filepath = '/kaggle/working/data.yaml'


with open(yaml_filepath, 'w') as f:
  doc = yaml.dump(
      dataset_info,
      f,
      default_flow_style=None,
      sort_keys=False
  )
```
