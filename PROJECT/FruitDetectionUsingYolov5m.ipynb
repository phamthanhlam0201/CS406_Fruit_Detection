```python
import yaml
import os

mapping = ['apple', 'orange', 'lemon',
           'avocado', 'cherry', 'coconut',
           'banana', 'watermelon', 'pineapple',
           'kiwi', 'tomato', 'mango']

dataset_info = {
    'train': 'dataset/train/images',
    'val': 'dataset/val/images',
    'test': 'dataset/test/images',
    'path': os.path.abspath('/content'),
    'nc': len(mapping),
    'names': mapping
}

yaml_filepath = '/content/dataset/data.yaml'


with open(yaml_filepath, 'w') as f:
  doc = yaml.dump(
      dataset_info,
      f,
      default_flow_style=None,
      sort_keys=False
  )
```

```python
!yolo train model=yolov5m.pt data='/content/dataset/data.yaml' epochs=50
```
